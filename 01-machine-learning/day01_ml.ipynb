```python
# X= Hours Studied
# y= Marks Obtained

X=[1,2,3,4,5]
y=[35,40,50,60,70]
```

```python
def simple_model(x):
    return 10*x+25
```

```python
predictions=[]
for i in X:
    predictions.append(simple_model(i))
print(predictions)
```

```python
errors=[]
for true,pred in zip(y,predictions):
    errors.append(abs(true-pred))
```

```python
print("Predictions:",predictions)
print("Errors:",errors)
```
